# MACHINE LEARNING: MÉTODO RIDGE

```python
#Librerias necesarias
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
#Carga de datos

datos = pd.read_csv('../data/kc_house_data.csv', sep = ',')
```

```python
datos.head(10)
```

# Puesta a punto de datos

```python
datos['date'] = pd.to_datetime(datos['date'],format = '%Y-%m-%d')
datos.loc[datos['yr_renovated'] != 0, 'yr_renovated'] = pd.to_datetime(datos.loc[datos['yr_renovated'] != 0, 'yr_renovated'], format = '%Y')

data_aux = datos

data_aux['yr_renovated_dummy'] = datos['yr_renovated'].apply(lambda x: 1 if x!=0 else 0)

data_aux = data_aux.drop(['yr_renovated'], axis=1)

data_aux['age_of_property'] = 2022 - datos['yr_built']


columns = ['price','bedrooms', 'bathrooms', 'sqft_living','waterfront', 'view', 'condition', 'grade', 'yr_renovated_dummy','age_of_property']

data_aux = data_aux[columns]

Y = data_aux['price']
X = data_aux.drop(columns = ['price'])

X.head()
```

# MODELO

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
ridge_model = Ridge(alpha=2, normalize = "True")
ridge_model 
```

```python
ridge_model.fit(X_train,Y_train)
```

```python
y_predict = ridge_model.predict(X_test)

print("RMSE: %.2f" % mean_squared_error(Y_test, y_predict, squared = False))

print("MSE: %.2f" % mean_squared_error(Y_test, y_predict, squared =True))

print("MAE: %.2f" % mean_absolute_error(Y_test, y_predict))

print("R^2: %.2f" % r2_score(Y_test, y_predict) )
```

```python
X_trainval, X_val, Y_trainval, Y_val = train_test_split(X_train, Y_train, test_size = 0.15, random_state = 0)
```

```python
ridge_model = Ridge(alpha=0.1, normalize = "True")
ridge_model.fit(X_trainval, Y_trainval)
y_predict1 = ridge_model.predict(X_val)
print("MAE: %.2f" % mean_absolute_error(Y_val, y_predict1))
```

```python
ridge_model = Ridge(alpha=0.5, normalize = "True")
ridge_model.fit(X_trainval, Y_trainval)
y_predict2 = ridge_model.predict(X_val)
print("MAE: %.2f" % mean_absolute_error(Y_val, y_predict2))
```

```python
ridge_model = Ridge(alpha=1, normalize = "True")
ridge_model.fit(X_trainval, Y_trainval)
y_predict5 = ridge_model.predict(X_val)
print("MAE: %.2f" % mean_absolute_error(Y_val, y_predict5))
```

```python
#El mejor valor de alpha es 0.1

ridge_model = Ridge(alpha=0.1, normalize = "True")
ridge_model.fit(X_train, Y_train)
y_predict_final = ridge_model.predict(X_test)

print("RMSE: %.2f" % mean_squared_error(Y_test, y_predict_final, squared = False))

print("MSE: %.2f" % mean_squared_error(Y_test, y_predict_final, squared =True))

print("MAE: %.2f" % mean_absolute_error(Y_test, y_predict_final))

print("R^2: %.2f" % r2_score(Y_test, y_predict_final) )

```
